## Import

```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path= "F:\Downloads\Documents\Cough Detector\Segmented Flusense Data\\train\Cough\\0_1JxnLOOUHaw_cough-160.wav" #0__-_5kbw2Mcw_cough-22.wav #
# file_path= "F:\Downloads\Documents\Cough Detector\Segmented Flusense Data\\train\Others\\0__0WKVY0n8aE_sneeze-13.wav"

f= 0 # for figure number
#loading file
signal, sr =  librosa.load(file_path, sr= 16000)  # signal is array of sample values data len= sr * T
plt.figure(f)
librosa.display.waveplot(signal, sr= sr)

# fft 
signal_fft= np.fft.fft(signal) 
signal_mag= np.abs(signal_fft) # as fft returns complex values
frequency= np.linspace(0, sr, int(len(signal_mag)))
f=f+1 
plt.figure(f), 
plt.plot(frequency, signal_mag)

#"""As fft of digital data is symmetrical we have to take half part of data for our analysis"""

left_frequency = frequency[:int(len(frequency)/2)]
signal_left_mag= signal_mag[:int(len(signal_mag)/2)]
f= f+1
plt.figure(f)
plt.plot(left_frequency, signal_left_mag)



## stft
n_fft= 400 # no of samples we will do fft for stft for each frame
hop_length= 200 #frame shift
signal_stft= librosa.core.stft(signal, n_fft= n_fft, hop_length= hop_length)

signal_stft_mag= np.abs(signal_stft)
f= f+1
plt.figure(f)
librosa.display.specshow(signal_stft_mag, sr= sr, hop_length= hop_length)
plt.colorbar()

db_signal_stft_mag= librosa.core.amplitude_to_db(signal_stft_mag)

f= f+1
plt.figure(f)
librosa.display.specshow(db_signal_stft_mag, sr= sr, hop_length= hop_length)
plt.colorbar()
```
